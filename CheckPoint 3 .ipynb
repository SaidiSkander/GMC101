```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv("titanic-passengers.csv",sep=';')
print (data.info())
print ('We Have '+str(data.isnull().sum().sum())+' Null Data')
data.head(10)
```

```python
print (data.isnull().sum())
data["Age"].fillna(data["Age"].mean(),inplace=True)
print (data.isnull().sum())

data.drop(["Cabin"],axis=1,inplace=True)


data.head()
```

```python
data.info()
data["Embarked"].fillna(data["Embarked"].describe()["top"],inplace=True)
```

```python
data.info()
data.head()
```

# At this point the Data is ready 

```python

from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
data['Sex']=enc.fit_transform(data['Sex'])
data['Survived']=enc.fit_transform(data['Survived'])


g=sns.FacetGrid(data,col='Survived',hue='Sex',aspect=1.7,height=10,legend_out = True)
g.map(plt.hist,'Age',bins=20) 
g.add_legend(title='Gender')
data.head()
```

```python
g=sns.FacetGrid(data,col='Survived',hue='SibSp',aspect=1.7,height=10,legend_out = True)
g.map(plt.hist,'Parch',bins=20) 
g.add_legend(title='Number Of SibSp')
data.head()
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map (data)
```
